```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Orignial_price = 90_000

purchase_price = 20_000 # variable 

sell_price_per = 0.5 # estimation
# action_cost = 0.18

sell_price = Orignial_price*sell_price_per  # 18% of platform is counted as price
transporting_cost = 500 # check it out
monthly_cost = 200 # not sure
month_till_sell = 12

tax = 0.25 # not sure
investors = 4
```

```python
def personal_profit(purchase_price, transporting_cost, monthly_cost, month_till_sell, sell_price_per):
    cost = purchase_price + monthly_cost*month_till_sell + transporting_cost
    profit = Orignial_price*sell_price_per
    return (profit-cost)*(1-tax)/investors

personal_profit(purchase_price, monthly_cost, transporting_cost, month_till_sell, sell_price_per)
```

```python
sim = 80000
sell_price_per = np.random.normal(0.5, 0.1, sim)
transporting_cost = np.random.normal(500, 50, sim)
monthly_cost = np.random.normal(200, 50, sim)
month_till_sell = np.random.normal(12, 2, sim)
```

```python
# Initialize 2D array to hold the histogram data
heatmap_data = []
bins = 20

for purchase_price in np.linspace(5000, 40_000, bins):
    profit = personal_profit(purchase_price, monthly_cost, transporting_cost, month_till_sell, sell_price_per)
    histogram_data, _ = np.histogram(profit, bins=bins, density=True )
    bins = _
    heatmap_data.append(histogram_data)

# Plotting the heatmap
plt.imshow(heatmap_data, aspect='auto', origin='lower', interpolation='bessel',
           extent=[bins[0], bins[-1], 5000, 40_000])  # 'extent' to match the axis values to our data
plt.colorbar(label='Frequency')
plt.ylabel('Purchase Price')
plt.xlabel('Profit')
plt.title('Profit Distribution Heatmap for Different Purchase Prices')
plt.show()
```
