```python
S = 5
N = 6

plt.title(f"Probability of winning {S=} {N=}")
for i,y in enumerate(all_dist(S, N)):
    plt.plot(range(1, N+1), y, label=f"Slot {i}")

#plt.yscale("log")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlabel("Number drawn N")
plt.ylabel("Probability")
plt.xticks(range(1, N+1))
plt.legend(loc='right')
plt.savefig(f"fig P4 20 M",dpi=300,bbox_inches="tight")
plt.show()


plt.title(f"Probability of winning {S=} {N=}")
for i,y in enumerate(all_dist(S,N).T):
    plt.plot(range(S), y, label=f"n {i}")
    # plt.scatter(range( 5), y)

#plt.yscale("log")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlabel("Chosen slot")
plt.ylabel("Probability")
plt.xticks(range(S))
plt.legend(loc='right')
plt.savefig(f"fig P3 20 M",dpi=300,bbox_inches="tight")
plt.show()
```
